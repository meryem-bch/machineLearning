```python
#Renommer la colonne default.payment.next.month
data.rename(columns={'default.payment.next.month':'DEFAULT'}, inplace=True)
```

```python
#supprimer les valeurs inutiles
data = data.drop(data[data['MARRIAGE']==0].index)
data = data.drop(data[data['EDUCATION']==0].index)
data = data.drop(data[data['EDUCATION']==5].index)
data = data.drop(data[data['EDUCATION']==6].index)
```

```python
#  graphique de distribution des classes basé sur la variable "default.payment.next.month"
plt.figure(figsize = (4,4))
graph = sns.countplot(x="DEFAULT", data=data, palette=['#3274a1','#e1812c'])

i=0
for p in graph.patches:
    height = p.get_height()  #obtenir la hauteur de chaque barre
    
    percentage = round(100 * data["DEFAULT"].value_counts()[i] / len(data),2) #calcul du pourcentage
    
    str_plot = f'{percentage} %'
    # ajouter le pourcentage au-dessus de chaque barre du graphique
    graph.text(p.get_x()+p.get_width()/2., height - 100, str_plot, ha="center")
    i += 1
    
plt.title('Class Distribution')
plt.xticks([0,1],['Non-defaulters', 'Defaulters'])
plt.xlabel('Default payment next month', fontsize=14)
plt.ylabel('Number of clients')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (14,6)) #créer un graphique avec une taille de figure spécifiée width: 14 et height: 6
plt.title('Amount of credit limit - Density Plot') #spécifier le titre de graphique
sns.histplot(data['LIMIT_BAL'],kde=True,bins=200, color="#3274a1") #créer un graphe de densité pour la colonne 'LIMIT_BAL' dans le DataFrame
plt.show()
```

```python
# Kernel density distribution
sns.set_palette(palette=['#3274a1','#e1812c'])
fig, axs = plt.subplots(1, 2, figsize=(15,4))

class0 = data.loc[data['DEFAULT'] == 0]['LIMIT_BAL']
class1 = data.loc[data['DEFAULT'] == 1]['LIMIT_BAL']
axs[0].set_title('Density plot of LIMIT_BAL by default type')
sns.kdeplot(class1, fill=True, label='Defaulters', ax=axs[0])
sns.kdeplot(class0, fill=True, label='Non-defaulters', ax=axs[0])
axs[0].axvline(0.14*1e6, 0, ls='--')  # ajouter une ligne verticale pointillée pour marquer une valeur spécifique sur l'axe des x
axs[0].legend() #la legende est ajouté a fin d ajouter identification de chaque classe

class0 = data.loc[data['DEFAULT'] == 0]['AGE']
class1 = data.loc[data['DEFAULT'] == 1]['AGE']
axs[1].set_title('Density plot of AGE by default type')
sns.kdeplot(class1, fill=True, label='Defaulters', ax=axs[1])
sns.kdeplot(class0, fill=True, label='Non-defaulters', ax=axs[1])
axs[1].axvline(26, 0, ls='--')
axs[1].axvline(40, 0, ls='--')
axs[1].legend()

fig.tight_layout()
plt.show()
```

```python
sns.set_palette(palette=['#3274a1','#e1812c'])
fig, axs = plt.subplots(1, 3, figsize=(20,4))

axs[0].set_title('SEX distribution')
sns.countplot(x='SEX', hue='DEFAULT', data=data, ax=axs[0])
axs[0].set_xticklabels(['Male', 'Female'])
axs[0].set_xlabel('')
axs[0].legend()

axs[1].set_title('EDUCATION distribution')
sns.countplot(x='EDUCATION', hue='DEFAULT', data=data, ax=axs[1])
axs[1].set_xticklabels(['Graduate School', 'University', 'High School', 'Others'])
axs[1].set_xlabel('')
axs[1].legend()

axs[2].set_title('MARRIAGE status distribution')
sns.countplot(x='MARRIAGE', hue='DEFAULT', data=data, ax=axs[2])
axs[2].set_xticklabels(['Married', 'Single', 'Others'])
axs[2].set_xlabel('')
axs[2].legend()

fig.tight_layout()
plt.show()
```

```python
payment_history = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
for col in payment_history:
    sns.countplot(x=col, hue='DEFAULT', data=data)
    plt.title(f'Default payment next month vs. {col}') 
    plt.show()
```
